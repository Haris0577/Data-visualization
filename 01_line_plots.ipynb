## Adding titles

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load dstaset
phool = sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phool ka plot")
plt.show()
```

## Adding limits

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load dstaset
phool = sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phool ka plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## Set styles
- darkgrid
- whitegrid
- dark
- white
- ticks

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style=None, rc=None)
#load dstaset
phool = sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phool ka plot")
sns.set_style("dark")
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style=None, rc=None)
#load dstaset
phool = sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phool ka plot")
sns.set_style("whitegrid")
plt.show()
```

## Size of figure


```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load dstaset
phool = sns.load_dataset("iris")
plt.figure(figsize=(6,8))
#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phool ka plot")
plt.show()
```
